```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("PrePlacement.csv")
data
```

```python
data.shape
```

```python
print("the number of rows in the dataset",data.shape[0])
print("the number of columns in the dataset",data.shape[1])
```

```python
import seaborn as sns
sns.jointplot(x='ssc_p',y='salary',data=data)
plt.show()
```

```python
import seaborn as sns
sns.jointplot(x='ssc_p',y='salary',data=data,kind='hex')
plt.show()
```

```python
#KDE-Kernel Density Distribution
import seaborn as sns
sns.jointplot(x='ssc_p',y='salary',data=data,kind='kde')
plt.show()
```

#  PairPlot

```python
sns.pairplot(data,hue='gender',diag_kind='kde',kind="scatter",palette='husl')
plt.show()
# Hue is used to differentiate
```

#  Categorical plot

```python
sns.stripplot(x='ssc_b',y='ssc_p',data=data)
plt.show()
```

```python
sns.stripplot(x='gender',y='ssc_p',data=data,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='gender',y='ssc_p',data=data)
plt.show()
# TRy to find about swarmplot in google and its uses
#Categorical vs numerical
```

# Distribution observation

```python
sns.violinplot(x='gender',y='ssc_p',data=data)
plt.show()
```

```python
sns.violinplot(x="specialisation",y="salary",data=data)
plt.show()
```

```python
sns.violinplot(x="specialisation",y="salary",data=data,hue='gender')
plt.show()
```

```python
sns.boxplot(x="ssc_b",y="ssc_p",data=data,hue='gender')
plt.show()
```

```python
sns.barplot(x="ssc_b",y="ssc_p",data=data,hue='gender')
plt.show()
```

```python
sns.countplot(x="ssc_b",data=data,hue='gender')
plt.show()
```

```python
sns.pointplot(x="ssc_p",y="ssc_b",data=data,hue='gender')
plt.show()
```
